```python
import pandas as pd
import ast
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data1 = pd.read_csv('Movie-Data.csv')
data2 = pd.read_csv('Movie.csv')
```

```python
data1.head()
```

```python
data2.head()
```

```python
data = pd.merge(data2,data1,on='title',how='inner')
```

```python
data.head(1)
```

```python
data = data[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.duplicated().sum()
```

# Data Preprossesing

```python
data.iloc[0].genres
```

```python
data.info()
```

```python
data['genres'][0]
```

```python
def conver_tag(x):
    l = []
    for i in (range(len(ast.literal_eval(x)))):
        l.append(ast.literal_eval(x)[i]['name'])
    return l
```

```python
data['genres'] = data['genres'].apply(conver_tag)
```

```python
data['genres']
```

```python
data['keywords'] = data['keywords'].apply(conver_tag)
```

```python
data['cast'][0]
```

```python
def conver_tag_cast(x):
    l = []
    cast = ast.literal_eval(x)
    for i in cast:
        l.append(i['name'])
    return l
```

```python
data['cast'] = data['cast'].apply(conver_tag_cast)
```

```python
data['cast']
```

```python
def crew_tag(x):
    l = []
    for i in ast.literal_eval(x):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
data['crew'] = data['crew'].apply(crew_tag)
```

```python
data['overview']
```

```python
data['overview'] = data['overview'].apply(lambda x:x.split(' '))
```

```python
data['overview']
```

```python
data.head(2)
```

```python
data['genres'] = data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['keywords'] = data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['cast'] = data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['crew'] = data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data.head(2)
```

```python
data['tags'] = data['overview'] + data['genres'] + data['cast'] + data['keywords'] + data['crew']
```

```python
data.head(3)
```

```python
data = data[['movie_id','title','tags']]
```

```python
data
```

```python
data['tags'] = data['tags'].apply(lambda x :' '.join(x))
```

```python
data
```

```python
data['tags'] = data['tags'].apply(lambda x:x.lower())
```

```python
data
```

```python
def stem(x):
    l = []
    for i in x.split(' '):
        l.append(PorterStemmer().stem(i))  # PorterStemmer().stem is function of nltk it used to remove similar word feature
        
    return " ".join(l)  # conver list into string
```

```python
data['tags'] = data['tags'].apply(stem)
```

```python
data
```

# Text-To-Vector

```python
cv = CountVectorizer(max_features=5000,stop_words='english')  # stop_word is used to remove word like "is,the,a,was,this"
vector = cv.fit_transform(data['tags']).toarray()             # max_feature is top 5000 repeated words
```

```python
vector
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

```python
len(cv.get_feature_names_out())
```

```python
similarity = cosine_similarity(vector)  # it masure distence with all row one by one like leave one out cross validation
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

# Final Output

```python
def recomend(movie):
    index = data[data['title'] == movie].index[0]  
    movie_list = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(data.iloc[i[0]].title)
```

```python
recomend("The Dark Knight Rises")      #Try Movie That Is Given Below 
```
